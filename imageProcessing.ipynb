```python
import cv2

# Read an image
image = cv2.imread("images/puppies.jpg")

if image is not None:
    print("Image loaded successfully!")
else:
    print("Failed to load image.")
```

```python
# Display the image
cv2.imshow("Displayed image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the image
cv2.imwrite("images/saved_image.jpg", image)
```

```python
# Resize an image by specifying exact dimension
new_width, new_height = 800, 600
resized_image = cv2.resize(image, (new_width, new_height))

# Display the image after resized by specifying exact dimension
cv2.imshow("Displayed image", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize an image by specifying scaling factor
scale_percent = 50
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)

resized_image_by_scale_factor = cv2.resize(image, (width, height))

# Display the image after resized by specifying scale factor
cv2.imshow("Diaplayed image", resized_image_by_scale_factor)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotate the image 90 degrees clockwise
center = (resized_image.shape[1]/2, resized_image.shape[0]/2)
angle = 90
scale = 1.0

rotated_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(resized_image, rotated_matrix, (resized_image.shape[1], resized_image.shape[0]))

cv2.imshow("Rotated image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Cropping an image
# Define the region of interest (ROI) coordinates
start_row, start_col = 100, 100 # Starting coordinated for ROI
end_row, end_col = 400, 400 # Ending coordinates for ROI

# Crop the image using the array slicing
cropped_image = resized_image[start_row:end_row, start_col, end_col]

cv2.imshow("Cropped image", cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
